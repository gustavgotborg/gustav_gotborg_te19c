```python
frukter = ["äpple", "päron", "kiwi", "banan"]
print(f"Fruktlista: {frukter}")
print(f"frukter{frukter[0]}")
print(f"frukt litan baklänges {frukter[::-1]}")
print("\n")

for frukt in frukter:
    print(f"Frukt: {frukt}")
```

```python
frukter = ["äpple", "päron", "kiwi", "banan"]
grönsaker = ["sallad", "majs", "tomat", "gurka"]

fruktsallad = []

for grönsak in grönsaker:
    fruktsallad.append(grönsak)

for frukt in frukter:
    fruktsallad.append(frukter)


print(fruktsallad)
```

```python
import matplotlib.pyplot as plt


y = [2*x-2 for x in range(10)]
x = [x for x in range(10)]
print(y)
print(x)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title(" 2*x-2")
```

```python
tabell = []

for i in range(6):
    rad = [x*i for x in range(6)]
    tabell.append(rad)
    for j in range(6):
        print(f"{tabell[i][j]}", end = " ")

    print()

print()

print(tabell)
print("\n")
print(tabell[3])
print(tabell[3][3])
```
