```python
#1 a)
import pandas as pd
import numpy as np
kf = pd.read_csv("sysselsattning.csv", encoding="ISO_8859-1", header=1)
kf_G = kf[kf["sysselsättning"]=="förvärvsarbetande"]
kf_sf = kf_G.sum(axis = 1).sum()
print(f"Antalet som är förvärvsarbetare är {kf_s}")
```

```python
#1 b)
import pandas as pd
import numpy as np
kf = pd.read_csv("sysselsattning.csv", encoding="ISO_8859-1", header=1)
kf_G = kf[kf["sysselsättning"]=="ej förvärvsarbetande"]
kf_se = kf_G.sum(axis = 1).sum()
print(f"Antalet som ej förvärvsarbetade är {kf_s}")
```

```python
#1 c) 
import matplotlib.pyplot as plt


arbete = 'förvärsarbetande', 'ej förförsarbetande'
förvärsarbetande = [kf_se, kf_sf]
 

plt.pie(förvärsarbetande, labels=arbete,
        autopct='%1.1f%%', colors= ["red", "blue"]) 
plt.axis('equal')
plt.show()
```
