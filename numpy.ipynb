```python
#3 
import numpy
import matplotlib.pyplot as plt
x = numpy.linspace(1,11)

def f(k, x, m):
    y = k*x + m 
    return y


plt.plot(f(3,x,1))
plt.title("3x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#4

import numpy
import matplotlib.pyplot as plt
x = numpy.linspace(-2,3)

def f(x, m):
    y = x**2 + m 
    return y


plt.plot(f(x,1))
plt.title("x^2 + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
